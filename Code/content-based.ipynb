```python
import string
import re
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
nltk.download('stopwords')
stop = set(stopwords.words('english'))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer, TfidfTransformer
```

```python
ratings_test = pd.DataFrame()
ratings_base = pd.DataFrame()
for i in range(1,53425):
    user_i = rating_new[rating_new['User-ID']==i]
    tenper = math.ceil(len(user_i)*0.1)
    tenper_user_i = user_i.tail(tenper)
    ratings_test = pd.concat([ratings_test, tenper_user_i], ignore_index = True)
    ratings_test.reset_index()
    nineper = len(user_i) - tenper
    nineper_user_i = user_i.head(nineper)
    ratings_base = pd.concat([ratings_base, nineper_user_i], ignore_index = True)
    ratings_base.reset_index()
rate_train = ratings_base.values
rate_test = ratings_test.values
```

```python
def remove_stopwords(text):
        if text is not None:
            tokens = [x for x in word_tokenize(text) if x not in stop]
            return " ".join(tokens)
        else:
            return None
def remove_punct(text):
    exclude = set(string.punctuation)
    s = ''.join(ch for ch in text if ch not in exclude)
    return s
def clean_text(data):
    df  = data
    df["dirty_title"] = data['book-title']
    df["book-title"] = df['book-title'].apply(lambda x : x.lower())
    df['book-title'] = df['book-title'].apply(lambda x : remove_stopwords(x)) 
    df['book-title'] = df['book-title'].apply(lambda x : remove_punct(x))
    ps = PorterStemmer()
    df['book-title'] = df['book-title'].apply(word_tokenize)
    df['book-title'] = df['book-title'].apply(lambda x: ' '.join([ps.stem(y) for y in x]))
    df['book-title'] = df['book-title'].replace('\s+', ' ', regex=True)
    return df
clean_title = clean_text(books_eng)[["isbn","book-title", "dirty_title"]]
```

```python
vect = CountVectorizer(analyzer = 'word', ngram_range = (1,1), stop_words = 'english', min_df = 0)
vect.fit(clean_title['book-title'])
title_matrix = vect.transform(clean_title['book-title'])
X_train_counts = pd.DataFrame(title_matrix.todense(), columns=vect.get_feature_names_out())
X_train_counts = X_train_counts.values
# title_matrix.shape
# X_train_counts
```

```python
X_train_counts
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
transformer = TfidfTransformer(smooth_idf=True, norm ='l2')
tfidf = transformer.fit_transform(X_train_counts.tolist()).toarray()
```

```python
import numpy as np
def get_items_rated_by_user(rate_matrix, user_id):
    y = rate_matrix[:,0] 
    ids = np.where(y == user_id+1)[0] 
    item_ids = rate_matrix[ids, 2] -1
    scores = rate_matrix[ids, 1]
    return (item_ids, scores)
```

```python
from sklearn.linear_model import Ridge
from sklearn import linear_model
n_users = 53425
d = tfidf.shape[1] 
W = np.zeros((d, n_users))
b = np.zeros((1, n_users))
for n in range(n_users):    
    ids, scores = get_items_rated_by_user(rate_train, n)
    clf = Ridge(alpha= 0.01, fit_intercept  = True)
    if(ids.size > 0):
        Xhat = tfidf[ids, :]
        
        clf.fit(Xhat, scores)  
        W[:, n] = clf.coef_  # ndarray of shape (n_features,) or (n_targets, n_features)
        b[0, n] = clf.intercept_ # float or ndarray of shape (n_targets,)
```

```python
Yhat = tfidf.dot(W) + b
```

```python
# Create function to calculate RMSE
def RMSE(Yhat,data):
    se = 0
    cnt = 0
    for n in range(n_users):
        ids, scores_truth = get_items_rated_by_user(data, n)
        scores_pred = Yhat[ids, n]
        e = scores_truth - scores_pred 
        se += (e*e).sum(axis = 0)
        cnt += e.size
    return np.sqrt(se/cnt)
```

```python
# Test with userId = 7
n = 11675
#np.set_printoptions(precision=2) # 2 digits after . 
ids, scores = get_items_rated_by_user(rate_test, n)
compareResult = {'book_id':ids,'True rate':scores,'Predicted':Yhat[ids,n]}
display(pd.DataFrame(compareResult))
```

```python
print("RMSE train: ",RMSE(Yhat,rate_train))
print("RMSE test: ",RMSE(Yhat,rate_test))
```
